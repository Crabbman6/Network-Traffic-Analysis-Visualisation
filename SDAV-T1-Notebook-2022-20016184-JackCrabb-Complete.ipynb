### Task 2: Plot a Line Chart that shows "Minutes" on the x-axis, and "Total Packet Length" sent on the y-axis. (3)

*Hint: Group you data by "Time" and then you can take the sum of the Length column.*

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx 
import seaborn as sns 
import numpy as np

data = pd.read_csv('./2022-task1_data.csv')
#What we need to do: Arithmetic operation on the "Time" Data Column 
#if time = 60, Minute + 1 
#if time = 120, Minute + 2 etc

#create the figure with size
plt.figure(figsize=(30,30))
#fig = figure
fig = plt.figure()

plt.axes(frameon=0)
#Divides Time column by 60 to turn into Minutes
data['Time'] = data['Time'].div(60)

data['Length'] = np.cumsum(data['Length'])

plt.xlabel("Minutes")
plt.ylabel("Total Packet Length")

plt.plot(data['Time'], data['Length'])
plt.show
```

### (Advanced) Task 6: Display a Node Link Diagram for this new DataFrame. (3)

*Hint: Look at the NetworkX library: https://networkx.org/ and the online course notes.*

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
import numpy as np

data = pd.read_csv('./2022-task1_data.csv')

sorted = data[data['Source'].str.startswith('10.')]
sorted

graph = nx.Graph()
graph = nx.from_pandas_edgelist(sorted, 'Source', 'Destination')

nx.draw(graph, with_labels = True)
plt.show()
```

### (Advanced) Task 8: Show a Multi-Line Chart that shows the Total Packet Length Per Protocol. (3)

*Hint: Think about how you did this in Task 1 and Task 2, and recall that plt.plot can be used to append to a plot.*

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
import numpy as np

data = pd.read_csv('./2022-task1_data.csv')
protocol_sums = data.groupby('Protocol')['Length'].sum()
plot = protocol_sums.plot(kind='line', title='Total Packet Length per Protocol')
plt.ylabel('Length')
plt.xticks(rotation=45)
plt.show()
```
